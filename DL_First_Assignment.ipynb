```
import numpy as np
import pandas as pd
```

```
dataset = pd.read_csv("50_Startups.csv")
```

```
#if label is missing then delete the record completely : As a best practice
```

```
#Delete all the records which is having the Nan
dataset.dropna(inplace=True)
#inplace= True it does not returns the data, But False will returns the data
```

```
#Rules for Regression
#1. Data should be complete
#2. Data should be in Numeric
#3. Features and label must be in the form form of 2D array
#4. For ANN Normalize the Feature and Label
#5. Features are normalized by Standardscalar or Robustscalar
#6. Labels are normalized with minMaxscalar
```

```
dataset.info()
```

```
dataset.corr()
```

```
#Now change the name to encoders
from sklearn import preprocessing
# State conversion
label_encoder = preprocessing.LabelEncoder()

encoded_state = label_encoder.fit_transform(dataset["State"])

dataset["State"] = encoded_state
```

```
features = dataset.drop("Profit",axis =1)
labels = dataset[["Profit"]]
features.shape,labels.shape
```

```
dataset.head()
```

```
#Feature scaling
from sklearn.preprocessing import  StandardScaler
std_sca=StandardScaler()
features_ss=std_sca.fit_transform(features)
features_ss
```

```
#label normalization
from sklearn.preprocessing import MinMaxScaler
min_max_sca = MinMaxScaler()
labels=min_max_sca.fit_transform(labels)
labels
```

```
#Spliting of data for training and testing
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(features_ss,labels,test_size=0.2,random_state=100)
X_train.shape, y_train.shape, X_test.shape, y_test.shape
```

```
#ANN Model 
import tensorflow as tf
#Create sequential model
model=tf.keras.models.Sequential()
```

```
#Create Dense layeres
model.add(tf.keras.layers.Dense(units=100, activation='sigmoid', input_shape=(4,)))

model.add(tf.keras.layers.Dense(units=100, activation='sigmoid'))

model.add(tf.keras.layers.Dense(units=1, activation='linear'))
```

```
#custom metric function
def r2_custom(y_true, y_pred):
  ss_res = tf.keras.backend.sum(tf.keras.backend.square(y_true - y_pred))

  ss_total = tf.keras.backend.sum(tf.keras.backend.square(y_true - tf.keras.backend.mean(y_true)))

  return (1-ss_res / (ss_total + tf.keras.backend.epsilon()))
```

```
# Compile the model
model.compile(optimizer="sgd" , loss="mean_squared_error", metrics=[r2_custom])
```

```
#History of data
history = model.fit(X_train,y_train,epochs=1000, validation_data=(X_test,y_test))
```

```
#trining score
model.evaluate(X_train,y_train)
```

```
#Testing score
model.evaluate(X_test, y_test) 
```

```
#Check the quality for genralized model or not
print("Training Score:", model.evaluate(X_train,y_train)[1])

print("Testing Score:", model.evaluate(X_test,y_test)[1])
```
